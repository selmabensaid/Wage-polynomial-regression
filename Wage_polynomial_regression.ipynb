```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df_class=df[['sex','maritl','race','education','region', 'jobclass','health','health_ins']].apply(le.fit_transform)
df_class.head()
df2=pd.concat([df.year, df.age, df_class, df.logwage,df.wage], axis=1)
```

# Preprocessing

```python
X=df2.drop(['wage', 'logwage'], axis=1)
```

```python
X.head()
```

```python
Y=df2.logwage
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

# Preprocessing

```python
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
def plot_polynomial(n):
    poly = PolynomialFeatures(n)
    X_poly = poly.fit_transform(X_train)
    X_test_poly=poly.fit_transform(X_test)
    poly.fit(X_poly,y_train)
    linear_reg=LinearRegression().fit(X_poly, y_train)
    plt.figure(figsize=(10,8))
    #plt.scatter(X,Y, color='r')
    sns.regplot(y_train, linear_reg.predict(X_poly), color='b')
    predictions=linear_reg.predict(X_test_poly)
    print('MAE:', metrics.mean_absolute_error(y_test, predictions))
    print('MSE:', metrics.mean_squared_error(y_test, predictions))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
    print('r2_score: ', metrics.r2_score(y_test, predictions))
    print ('r2_adjusted',  1 - (1- metrics.r2_score(y_test, predictions))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
```

# Lasso Polynomial regression

```python
from sklearn.linear_model import LassoCV
```

```python
def plot_polynomial_LassoCV(n):
    poly = PolynomialFeatures(n)
    X_poly = pd.DataFrame(poly.fit_transform(X_train))
    X_test_poly=poly.fit_transform(X_test)
    poly.fit(X_poly,y_train)
    lasso_reg=LassoCV().fit(X_poly, y_train)
    lasso_reg.fit(X_poly, y_train)
    print("Best alpha using built-in LassoCV: %f" % lasso_reg.alpha_)
    print("Best score using built-in LassoCV: %f" %lasso_reg.score(X_test_poly,y_test))
    coef = pd.Series(lasso_reg.coef_, index = X_poly.columns)
    print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
    imp_coef = coef.sort_values()
    #plt.rcParams['figure.figsize'] = (8.0, 10.0)
    #imp_coef.plot(kind = "barh")
```

# Polynomial regression using the wage alone

```python
age=df2['age']

y=df2.wage
```

```python
from sklearn.model_selection import train_test_split
age_train, age_test, y_train, y_test = train_test_split(age, y, test_size=0.3)
age_train_original=age_train
age_train=np.array(age_train).reshape(-1,1)
age_test=np.array(age_test).reshape(-1,1)
```

```python
def plot_polynomial_age(n):
    poly = PolynomialFeatures(n)
    X_poly = poly.fit_transform(age_train)
    X_test_poly=poly.fit_transform(age_test)
    poly.fit(X_poly,y_train)
    linear_reg=LinearRegression().fit(X_poly, y_train)
    plt.figure(figsize=(10,8))
    #plt.scatter(X,Y, color='r')
    plt.scatter(age_train_original, y_train, color='r')
    plt.scatter(age_train_original, linear_reg.predict(X_poly), color='b')
   
    predictions=linear_reg.predict(X_test_poly)
    print('MAE:', metrics.mean_absolute_error(y_test, predictions))
    print('MSE:', metrics.mean_squared_error(y_test, predictions))
    print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
    print('r2_score: ', metrics.r2_score(y_test, predictions))
    print ('r2_adjusted',  1 - (1- metrics.r2_score(y_test, predictions))*(len(y_train)-1)/(len(y_train)-age_train.shape[1]-1))
```

```python
plot_polynomial_age(2)
```

```python
plot_polynomial_age(3)
```
