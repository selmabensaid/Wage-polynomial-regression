```python
X=np.array(df.Level).reshape(-1, 1)
Y=df.Salary
```

```python
# Import
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, Y)

# calculate SVM Regressor model metrics
# from sklearn import metrics
# print('Linear Regression:')
# print('MAE::', metrics.mean_squared_error(Y, predictions))
# print('RMSE:', np.sqrt(metrics.mean_squared_error(Y, predictions)))
# print('r2_score: ', metrics.r2_score(Y, predictions))
# print ('r2_adjusted',  1 - (1- metrics.r2_score(Y, predictions))*(len(Y)-1)/(len(Y)-X.shape[1]-1))
```

```python
plt.scatter(X,Y, color='r')
plt.plot(X, reg.predict(X), color='b')
```

```python
from sklearn import metrics
```

```python
def plot_polynomial(n):
    poly = PolynomialFeatures(n)
    X_poly = poly.fit_transform(X)
    poly.fit(X_poly,Y)
    linear_reg=LinearRegression().fit(X_poly, Y)
    plt.figure(figsize=(10,8))
    plt.scatter(X,Y, color='r')
    plt.plot(X, linear_reg.predict(X_poly), color='b')
    print('r2_score: ', metrics.r2_score(Y,linear_reg.predict(X_poly)))
    
    
```

```python
plot_polynomial(2)
```

```python
plot_polynomial(3)
```

```python
plot_polynomial(4)
```
